```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df=pd.read_csv('income.csv')
df.head()
```

```python
plt.scatter(df.Age,df[['Income($)']])
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
km
```

```python
df['predicted']=km.fit_predict(df[['Age','Income($)']])
df
```

```python
plt.scatter(df[df.predicted==0]['Age'],df[df.predicted==0]['Income($)'],color='red',marker='+')
plt.scatter(df[df.predicted==1]['Age'],df[df.predicted==1]['Income($)'],color='blue',marker='+')
plt.scatter(df[df.predicted==2]['Age'],df[df.predicted==2]['Income($)'],color='green',marker='+')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

df['Income($)']=scaler.fit_transform(df[['Income($)']])
df.Age=scaler.fit_transform(df[['Age']])
df
```

```python
kmeans=KMeans(n_clusters=4)
df['Predicted2']=kmeans.fit_predict(df[['Age','Income($)']])
df
```

```python
plt.scatter(df[df.predicted==0]['Age'],df[df.predicted==0]['Income($)'],color='red',marker='+')
plt.scatter(df[df.predicted==1]['Age'],df[df.predicted==1]['Income($)'],color='blue',marker='+')
plt.scatter(df[df.predicted==2]['Age'],df[df.predicted==2]['Income($)'],color='green',marker='+')
plt.scatter(df[df.predicted==3]['Age'],df[df.predicted==3]['Income($)'],color='black',marker='+')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='*')
```

```python
kmeans.cluster_centers_
```

```python
ss=[]
for i in range(1,10):
    km=KMeans(n_clusters=i)
    km.fit(df[['Age','Income($)']])
    ss.append(km.inertia_)
plt.plot(range(1,10),ss)
```
