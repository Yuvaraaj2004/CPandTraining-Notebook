```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
```

```python
df = load_iris()

df, target = pd.DataFrame(df.data, columns=df.feature_names), pd.DataFrame(
    df.target, columns=['target'])
df,target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test =train_test_split(df,target,test_size=0.2,random_state=10)

```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
X_train
```

```python
from sklearn.metrics import accuracy_score


pred=model.predict(X_test)
print({'predict':pred,'actual':y_test.to_numpy().reshape(1,len(y_test))})
accuracy_score(y_test,pred)
ndf=pd.DataFrame({'predict':pred,'actual':y_test.to_numpy().reshape(1,len(y_test))[0]})
ndf
```
