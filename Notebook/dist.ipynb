```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
plt.plot([5, 6, 8, 5, 6], [8, 8, 12, 9, 11])
```

```python

model = LinearRegression().fit(
    [[i] for i in [5, 6, 8, 5, 6]], [8, 8, 12, 9, 11])
```

```python
model.intercept_, model.coef_
```

```python
plt.plot([6.89, 5.12, 7.82, 7.42, 6.94, 7.89, 6.73, 6.75, 6.09],
         [3.26, 1.98, 3.25, 3.67, 3.57, 2.99, 2.6, 2.48, 2.31])

plt.scatter(x=[6.89, 5.12, 7.82, 7.42, 6.94, 7.89, 6.73, 6.75, 6.09],
            y=[3.26, 1.98, 3.25, 3.67, 3.57, 2.99, 2.6, 2.48, 2.31], marker='*', c='r')
```

```python
import numpy
y = numpy.array([3.26, 1.98, 3.25, 3.67, 3.57,
                2.99, 2.6, 2.48, 2.31]).T
x = numpy.array([6.89, 5.12, 7.82, 7.42, 6.94,
                7.89, 6.73, 6.75, 6.09], ndmin=2).T
print(x.shape, y.shape)
model = LinearRegression()
model.fit(x,
          y)
```

```python
model.coef_, model.intercept_
```

```python

model.predict([[6.89]])
```

```python
df = pd.read_csv('../csv/insurance_data.csv')
df.head()
```

```python
plt.scatter(df.age, df.bought_insurance)
```

```python
from sklearn.linear_model import LogisticRegression


model = LinearRegression()
model.fit(df[['age']], df.bought_insurance)
LogismModel = LogisticRegression()
LogismModel.fit(df[['age']], df.bought_insurance)
```

```python
print(model.coef_, model.intercept_)
LogismModel.predict(df[['age']])
```
