```python
# Import the required libraries
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
import pandas as pd
pd.set_option('display.max_columns', 100)
warnings.simplefilter('ignore')

# Load the dataset
data = pd.read_csv("../csv/iris.csv")
```

```python
def func(x, y, title, filename, ref,):
    try:
        plt.axis('equal')
        ref(data=data, x=x, y=y)
        plt.title(title)
        plt.savefig(filename)
    except Exception as e:
        print(e,title)
```

```python
for i in [
    ['petal_length', 'petal_width', 'num_lineplot','num_lineplot.png', sns.scatterplot],
    ['petal_length', 'petal_width', 'num_scatterplot','num_scatterplot.png', sns.scatterplot],
    ['petal_length', 'petal_width', 'num_regplot', 'num_regplot.png', sns.regplot],
    ['petal_length', 'petal_width', 'num_regplot','num_jointplot.png', sns.jointplot],
    ['species_type', 'petal_width', 'cantum_barplot','cantom_barplot.png', sns.barplot],
    ['species_type', 'petal_width', 'cantum_pointplot','cantom_pointplot.png', sns.pointplot],
    ['species_type', 'petal_width', 'cantum_boxplot','cantom_boxplot.png', sns.boxplot],
    ['species_type', 'petal_width', 'cantum_violinplot','cantom_violinplot.png', sns.violinplot],
    ['species_type', 'petal_width', 'cantum_stripplot','cantom_stripplot.png', sns.stripplot],
    ['species_type', 'petal_width', 'cantum_swarmplot','cantom_swarmplot.png', sns.swarmplot],
]:
    func(*i)
```
