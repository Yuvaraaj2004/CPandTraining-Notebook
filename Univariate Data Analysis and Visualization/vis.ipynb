```python
# Import the required libraries
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
import pandas as pd
pd.set_option('display.max_columns', 100)
warnings.simplefilter('ignore')
```

```python
data = pd.read_csv("vehicle_data_1.csv")
```

```python
print("DF Shape")
data.shape
```

```python
num_cols = data.select_dtypes(exclude='object')
num_cols
```

```python
def ploting(data,title,num_cols,ref,fileName,rows=3,cols=3,hs=0.5,ws=0.5,):
    fig=plt.figure(figsize=(cols*5,rows*5))
    plt.subplots_adjust(hspace=hs,wspace=ws)
    plt.suptitle(title)
    for i,col in enumerate(num_cols):
        ref(x=data[col],ax=fig.add_subplot(rows,cols,i+1))
    plt.savefig(fileName)
```

```python
l=[
    {'data':data,'title':'histogram','num_cols':num_cols,'ref':sns.histplot,'fileName':'histogram.png'},
    {'data': data, 'title': 'KDE plot', 'num_cols': num_cols, 'ref': sns.kdeplot, 'fileName': 'kde.png'},
    {'data': data, 'title': 'Rug plot', 'num_cols': num_cols, 'ref': sns.rugplot, 'fileName': 'rugplot.png'},
    {'data': data, 'title': 'Box plot', 'num_cols': num_cols, 'ref': sns.boxplot, 'fileName': 'boxplot.png'},
    {'data': data, 'title': 'Violin plot', 'num_cols': num_cols, 'ref': sns.violinplot, 'fileName': 'violinplot.png'},
    {'data': data, 'title': 'Strip plot', 'num_cols': num_cols, 'ref': sns.stripplot, 'fileName': 'stripplot.png'},
    {'data': data, 'title': 'count plot', 'num_cols': data.select_dtypes(include='object'), 'ref':sns.countplot, 'fileName': 'countplot.png','rows':2,'cols':2},
    {'data': data, 'title': 'pie chart', 'num_cols': data.select_dtypes(include='object'), 'ref': plt.pie, 'fileName': 'piechart.png', 'rows': 3, 'cols': 3}]
for i in l:
    ploting(**i)
```
